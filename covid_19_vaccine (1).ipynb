#extracting important columns for us, on which we are going to work

```python
data1 = data1.rename(columns={'Updated On':'date', 'State':'state','Male(Individuals Vaccinated)':'male_vaccinated',
       'Female(Individuals Vaccinated)':'female_vaccinated', 'Transgender(Individuals Vaccinated)':'trans_vaccinated','Total Doses Administered':'total_doses'})
```

```python
data1 = data1[["date","state","male_vaccinated","female_vaccinated","trans_vaccinated","total_doses"]]
```

```python
#fill NaN values with precedind values to make it nuetral
data1 = data1.fillna(method = 'ffill')
```

# Analysing our data and plotting graph for that

```python
#temp = data1[data1.date == "2021-3-05"]
temp_state = data1.state.unique()
```

```python
new_temp_state = ipywidgets.Dropdown(
                options=temp_state,
                #options = temp_value
                value=temp_state[0],
                description='Select state:',
                disabled=False,
                )
def select (select):
    temp2_state = data1[data1.state == new_temp_state.value]
    temp_value = ["total_doses","female_vaccinated","trans_vaccinated","male_vaccinated"]
    data_select2 = ipywidgets.Dropdown(
                options= temp_value,
                value=temp_value[0],
                description='Select option:',
                disabled=False,
                )
    
    #print("first",new_temp_state)
    #print("second",temp2_state)
    def data_select(data_select):
        
        sns.set(rc = {'figure.figsize':(15,10)})
        if data_select2.value == "total_doses":
            plot1 = sns.lineplot(x = "date",y = "total_doses", data = temp2_state,color = "g")
        elif data_select2.value == "female_vaccinated": 
            plot2 = sns.lineplot(x = "date",y = "female_vaccinated", data = temp2_state, color = "b")
        elif data_select2.value =="trans_vaccinated":
            plot3 = sns.lineplot(x = "date",y = "trans_vaccinated", data = temp2_state, color = "r")
        elif data_select2.value =="male_vaccinated":
            plot3 = sns.lineplot(x = "date",y = "male_vaccinated", data = temp2_state, color = "y")
        #plt.title(new_temp_state)
    data_select(data_select2)
    ipywidgets.interact(data_select, data_select = data_select2)
```

```python
# Here Green Line indicating  Total doses of Covid vaccination in that state or country
# Here Blue Line indicating  female vaccination of Covid vaccination in that state or country
# Here Red Line indicating  Transgender vaccination of Covid vaccination in that state or country
# Here yellow Line indicating  male vaccination of Covid vaccination in that state or country
```

# Here we are predicting Total vaccination in future

```python
#using ipywidgets library to create dropdown option for user to easily select the query
temp_value = ["total_doses","female_vaccinated","trans_vaccinated","male_vaccinated"]
data_select3 = ipywidgets.Dropdown(
            options= temp_value,
            value=temp_value[0],
            description='Select option:',
            disabled=False,
            )
#making our ultimate state predictor function to predict our data statewise
def ultimate_prediction(predict):
    #creating datepicker for user
    select_date = ipywidgets.DatePicker(
    description='Pick a Date',
    disabled=False
    )
    #creating predict_data function for returning the future cases depend upon the user selection
    def predict_data(prediction):
        from sklearn.model_selection import train_test_split
        #print("first",new_temp_state)
        temp_state_value = data1[data1.state == new_temp_state.value]
        temp_state_value['date'] = temp_state_value['date'].map(dt.datetime.toordinal)
        #print("second",temp_state_value['date'])
        #temp_state_value.head()
        x = temp_state_value['date']
        temp = data_select3.value
        #print(temp)
        y = temp_state_value[temp]
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
        from sklearn.ensemble import RandomForestRegressor
        rf = RandomForestRegressor()
        rf.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
        #print(select_date.value)
        choose_date = select_date.value
        x = None
        #print(choose_date)
        choose_date2 = dt.date.today()
        
        choose_date2 = choose_date2.toordinal()
        #print(choose_date)
        if choose_date == x:
            result = rf.predict([[choose_date2]])
        if choose_date != x:
            choose_date = choose_date.toordinal()
            result = rf.predict([[choose_date2]])     
        return result
    #predict_data(prediction)
    ipywidgets.interact(predict_data, prediction = select_date)
new_temp_state
    
```

```python
ipywidgets.interact(ultimate_prediction, predict = data_select3)
```
